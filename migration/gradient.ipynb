```python
from src.plots import env 
from src.plots.plots.utils import named_colors 
import vice 
import matplotlib.pyplot as plt 
```

```python
out = vice.output("./outputs/high-resolution/diffusion/lateburst") 
```

```python
fig = plt.figure(figsize = (7, 7)) 
ax = fig.add_subplot(111, facecolor = "white") 
ax.set_xlabel("[Fe/H]") 
ax.set_ylabel("MDF") 
ax.set_xlim([-0.6, 0.6]) 
ax.set_ylim([0, 10]) 
# stem = "./figures/MDFs_4_5_kpc" 
for i in range(10, 20): 
    mode_idx = out.zones["zone%d" % (i)].mdf["dn/d[fe/h]"].index(
        max(out.zones["zone%d" % (i)].mdf["dn/d[fe/h]"])) 
    mode = (out.zones["zone%d" % (i)].mdf["bin_edge_left"][mode_idx] + 
            out.zones["zone%d" % (i)].mdf["bin_edge_right"][mode_idx]) / 2. 
    ax.plot(2 * [mode], [0, 10], c = named_colors()["black"], linestyle = ':')  
    xvals = [(a + b) / 2. for a, b in zip(out.zones["zone%d" % (i)].mdf["bin_edge_left"], 
                                         out.zones["zone%d" % (i)].mdf["bin_edge_right"])] 
    ax.plot(xvals, out.zones["zone%d" % (i)].mdf["dn/d[fe/h]"]) 
# for i in ["pdf", "png"]: plt.savefig("%s.%s" % (stem, i)) 
```
